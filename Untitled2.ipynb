```python
def get_playlist_tracks(playlist_id):

    results = sp.user_playlist_tracks("spotify", playlist_id)
    tracks = results['items']

    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])

    return tracks

tracks = get_playlist_tracks("12Wbv8sIx84T5uh6iOoJ7V")
```

```python
track_ids = [track['track']['uri'] for track in tracks]
```

```python
def get_artists_ids_from_playlist(playlist_id):

    tracks_from_playlist = get_playlist_tracks(playlist_id)

    artists = []
    artists_ids = []

    for track in tracks_from_playlist:
        artists_info = track['track']['artists']

        for artist_info in artists_info:
            artists.append(artist_info['name'])
            artists_ids.append(artist_info['id'])

    # artists = [(artist["name"], artist["id"]) for track in tracks_from_playlist for artist in track['track']['artists']]

    return list(set(artists_ids))

artists_ids = get_artists_ids_from_playlist("12Wbv8sIx84T5uh6iOoJ7V")
len(artists_ids)
```

```python
# artist_top_tracks endpoint returns 10 top songs of the artist
# BUT it only returns the songs as albums
# even if there's single track in the top list it considers it as an album with one track
artists_ids = artists_ids[:100]
album_ids = []
for artist_id in artists_ids:
    top_tracks = sp.artist_top_tracks(artist_id)
    album_ids.extend([track["album"]["uri"] for track in top_tracks['tracks']])

album_ids = album_ids[:200]
```

```python
all_track_ids = []
for album_id in album_ids:
    album_tracks = sp.album_tracks(album_id)
    for track in album_tracks['items']:
        all_track_ids.append(track['uri'])

print(len(all_track_ids))
```

```python
for track in tracks:
    all_track_ids.append(track["track"]["uri"])
    
print(all_track_ids)
```

```python
import pandas as pd

start = 0
dataframe = pd.DataFrame()

for stop in range(0, len(all_track_ids) + 100, 100):
    if start != stop:
        audio_features = sp.audio_features(all_track_ids[start:stop])
        new_dataframe = pd.json_normalize(audio_features)
        dataframe = pd.concat([new_dataframe, dataframe])
        start = stop

dataframe.reset_index()
```
